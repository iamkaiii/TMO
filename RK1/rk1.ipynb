Корреляционный анализ и скрипичная диаграмма для Admission_Predict.csv

1. Проверка и удаление пропущенных значений
Сначала я проверил наличие пропущенных значений в датасете:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
data = pd.read_csv('Admission_Predict.csv')

# Проверка пропущенных значений
print(data.isnull().sum())
```

2. Корреляционный анализ
Я провел корреляционный анализ, чтобы понять взаимосвязи между признаками:

```python
# Вычисление корреляционной матрицы
corr_matrix = data.corr(numeric_only=True)

# Визуализация корреляционной матрицы
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

3. Скрипичная диаграмма (Violin plot)
Я построил скрипичную диаграмму для признака "CGPA":

```python
plt.figure(figsize=(10, 6))
sns.violinplot(y='CGPA', data=data, inner='quartile', palette='Set2')
plt.title('Violin Plot for CGPA')
plt.ylabel('CGPA Score')
plt.show()  
```
